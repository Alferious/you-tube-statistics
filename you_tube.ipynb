```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None) # This displays all columns in the dataset
df = pd.read_csv("global_youtube_statistics.csv")
```

```python
df = df[['rank', 'Youtuber', 'subscribers', 'video views','category','uploads','Country','channel_type','video_views_rank','country_rank','channel_type_rank','lowest_monthly_earnings','highest_monthly_earnings',
        'lowest_yearly_earnings','highest_yearly_earnings','created_year','Population']] #defining all the columns which I will use
```

```python
# Viewing the first 5 rows of the data
df.head()
```

```python
df.info() # gives the summary of the data
```

```python
# Checking if there are null data
df.isnull().sum()
```

The data has the folowing columns having null values: category, Country, Channel_type, video_views_rank, country_rank, channel_type, channel_type_rank, created_year, population

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df['channel_type'].unique() #identifying unique channel types in the data
```

```python
grouped_data = df.groupby(by=['channel_type'])['subscribers'].sum().reset_index()
sorted_data = grouped_data.sort_values(by='subscribers', ascending=False)
sorted_data
```

```python
import seaborn as sns
```

```python
sns.lineplot(x=sorted_data['channel_type'], y=sorted_data['subscribers'])
plt.title('Channel Type Statisctics')
plt.xticks(rotation=85)
plt.show()
```

```python
earnings = df.groupby('channel_type').agg(
    highest_earning = ('highest_yearly_earnings', 'max'),
    lowest_earning = ('lowest_yearly_earnings', 'min')
).reset_index()

sorted_earnings = earnings.sort_values(by='highest_earning', ascending=False)

sorted_earnings
```

```python
sns.lineplot(x = sorted_earnings['channel_type'],y= sorted_earnings['highest_earning'])
plt.title('Youtube Yearly Earnings')
plt.xlabel('channel_type')
plt.ylabel('Earnings')
plt.xticks(rotation=85)
plt.show()


```

```python
earnings = df.groupby('channel_type').agg(
    highest_earning = ('highest_monthly_earnings', 'max'),
    lowest_earning = ('lowest_monthly_earnings', 'min')
).reset_index()

sorted_earning = earnings.sort_values(by='highest_earning', ascending=False)
sorted_earning
```

```python
sns.lineplot(x = sorted_earning['channel_type'],y= sorted_earning['highest_earning'])
plt.title('Youtube Monthly Earnings')
plt.xlabel('channel_type')
plt.ylabel('Earnings')
plt.xticks(rotation=85)
plt.show()
```

```python
uploads = df.groupby(by = ['Youtuber', 'subscribers'])['uploads'].sum().reset_index()

uploads = uploads.sort_values(by=['uploads'], ascending=False)
uploads.head(10)
```

```python
sns.lineplot(x = uploads.head(10)['Youtuber'],y= uploads.head(10)['uploads'])
plt.title('Youtuber uploads')
plt.xlabel('Youtuber')
plt.ylabel('uploads')
plt.xticks(rotation=85)
plt.show()
```

```python
views = df.groupby(by = ['Youtuber', 'subscribers'])['video views'].sum().reset_index()

views = views.sort_values(by=['video views'], ascending=False)
views.head(10)
```

```python
sns.lineplot(x = views.head(10)['Youtuber'],y= views.head(10)['video views'])
plt.title('You tuber viewership')
plt.xlabel('Youtuber')
plt.ylabel('video views')
plt.xticks(rotation=85)

plt.show()
```
